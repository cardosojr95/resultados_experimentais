# Gráfico em escala linear de um diodo associado com resistência em série

```python
# importando bibliotecas
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.font_manager import FontManager

#diretorio
caminho =r"C:\Users\cardo\Jupyter Notebook\dados_experimentais\Serie"

#arquivos do diretorio
dados = ["\diodo_rs_10ohm.xlsx","\diodo_rs_150ohm.xlsx","\diodo_rs_510ohm.xlsx"]

# valores de resistências
rs = ['$R_{s}= 10\Omega$', '$R_{s}= 150\Omega$', '$R_{s}= 510\Omega$']

# lista vazia
dataframes_list = []
  
# adicionando dados na lista vazia
for i in range(len(dados)):
    temp_df = pd.read_excel(caminho+dados[i])
    dataframes_list.append(temp_df)

# reorganizando os dados dos datasets
for j in range(len(dataframes_list)):
    dataframes_list[j].sort_values(by = 'Voltagem(V)',ascending = True, inplace = True)
    dataframes_list[j].sort_values(by = 'Corrente(A)',ascending = True, inplace = True)

#construção dos gráficos
for i in range(len(dataframes_list) and len(rs)):
    plt.plot(dataframes_list[i]['Voltagem(V)'], dataframes_list[i]['Corrente(A)'],label = rs[i])

    
plt.xlabel('Voltagem (V)')
plt.ylabel('Corrente (nA)')
plt.title('Representação em escala linear')
plt.grid(linestyle = '--',linewidth = 0.5)
plt.legend()

```
